## Cleaning balanced flights sample

```python
path = '/Users/reneehall/Documents/Learning/lighthouse_labs/coursework-lighthouse-labs/Mid-term project/colab_data/'
sample_df = pd.read_csv(path+'balanced_sample.csv')
```

```python
sample_df.shape
```

```python
# missing data for flights_sample
total = sample_df.isnull().sum().sort_values(ascending=False)
percent = (sample_df.isnull().sum()/sample_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(15)
```

```python
# look at missing categorical data
cat_cols = sample_df.select_dtypes(include='object')
(cat_cols.isnull().sum()/cat_cols.isnull().count()).sort_values(ascending=False)
```

```python
# look at missing continuous data
var_cols = sample_df.drop(labels=cat_cols.columns.tolist(), axis=1)
(var_cols.isnull().sum()/var_cols.isnull().count()).sort_values(ascending=False)
```

#### Cleaning cat_cols

```python
# drop cancellation_code, too many null values
sample_df = sample_df.drop(labels='cancellation_code', axis=1)
```

```python
# replace tail_num NaN's with 'tail_num_unknown'
sample_df['tail_num'] = sample_df['tail_num'].fillna('tail_num_unknown')
```

#### Cleaning var_cols

```python
# will keep the delay time columns (ex. weather_delay etc) in case we make it to stretch work
# will need to drop when we load datasets for initial model building 
# run below when loading dataset to drop these:
#to_drop = ['first_dep_time','longest_add_gtime','total_add_gtime','late_aircraft_delay','weather_delay','carrier_delay','security_delay','nas_delay']
#sample_df = sample_df.drop(labels=to_drop, axis=1)
```

```python
# replace remaining null values with mean
to_mean = ['air_time','arr_delay','actual_elapsed_time','taxi_in','wheels_on','wheels_off','arr_time','taxi_out','dep_delay','dep_time','crs_elapsed_time']
sample_df[to_mean].head()
```

```python
for col in to_mean:
    sample_df[col] = sample_df[col].fillna(sample_df[col].mean())
```

```python
# the optional stretch columns still have NaN's but remaining columns that will be used in inital models are cleaned
(sample_df.isnull().sum()/sample_df.isnull().count()).sort_values(ascending=False).head(20)
```

```python
sample_df.shape
```

```python
# save cleaned sample
sample_df.to_csv(path+'cleaned_balanced_sample.csv', index=False)
```

## Cleaning fuel_consumption table

```python
# load fuel consumption table
path = '/Users/reneehall/Documents/Learning/lighthouse_labs/coursework-lighthouse-labs/Mid-term project/colab_data/'
fuel_df = pd.read_csv(path + 'fuel_comsumption.csv')
fuel_df.head()
```

```python
# missing data for fuel
total = fuel_df.isnull().sum().sort_values(ascending=False)
percent = (fuel_df.isnull().sum()/fuel_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(10)
```

```python
# isolate categorial and continuous columns
cat_cols = fuel_df.select_dtypes(include='object')
var_cols = fuel_df.drop(labels=cat_cols.columns.tolist(), axis=1)
```

#### Cleaning cat_cols

```python
# check for missing values
(cat_cols.isnull().sum()/cat_cols.isnull().count()).sort_values(ascending=False)
```

```python
# replace nan's with unknown category
fuel_df['carrier_name'] = fuel_df['carrier_name'].fillna('carrier_name_unknown')
fuel_df['unique_carrier'] = fuel_df['unique_carrier'].fillna('unique_carrier_unknown')
```

#### Cleaning var_cols

```python
# check for missing values
(var_cols.isnull().sum()/var_cols.isnull().count()).sort_values(ascending=False)
```

```python
# replace Nan's with mean
to_mean = ['total_gallons','tdomt_gallons','slat_gallons','tint_gallons','spac_gallons','tint_cost','airline_id','sint_cost','slat_cost','spac_cost','satl_cost','sdomt_gallons','satl_gallons','ts_gallons','sint_gallons']
for col in to_mean:
    fuel_df[col] = fuel_df[col].fillna(fuel_df[col].mean())
```

```python
# save to csv
fuel_df.to_csv(path+'cleaned_fuel_consumption.csv', index=False)
```
