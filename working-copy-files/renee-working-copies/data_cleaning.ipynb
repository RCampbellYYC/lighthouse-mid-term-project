## fuel_comsumption

```python
path = '/Users/reneehall/Documents/Learning/lighthouse_labs/coursework-lighthouse-labs/Mid-term project/colab_data/'
fuel_df = pd.read_csv(path + 'fuel_comsumption.csv')
fuel_df.head()
```

```python
# missing data for fuel
total = fuel_df.isnull().sum().sort_values(ascending=False)
percent = (fuel_df.isnull().sum()/fuel_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(10)
```

```python
cat_cols = fuel_df.select_dtypes(include='object')
var_cols = fuel_df.drop(labels=cat_cols.columns.tolist(), axis=1)
```

#### Cleaning cat_cols

```python
(cat_cols.isnull().sum()/cat_cols.isnull().count()).sort_values(ascending=False)
```

```python
# replace nan's with unknown category
fuel_df['carrier_name'] = fuel_df['carrier_name'].fillna('carrier_name_unknown')
fuel_df['unique_carrier'] = fuel_df['unique_carrier'].fillna('unique_carrier_unknown')
```

#### Cleaning var_cols

```python
(var_cols.isnull().sum()/var_cols.isnull().count()).sort_values(ascending=False)
```

```python
# replace Nan's with mean
to_mean = ['total_gallons','tdomt_gallons','slat_gallons','tint_gallons','spac_gallons','tint_cost','airline_id','sint_cost','slat_cost','spac_cost','satl_cost','sdomt_gallons','satl_gallons','ts_gallons','sint_gallons']
for col in to_mean:
    fuel_df[col] = fuel_df[col].fillna(fuel_df[col].mean())
```

```python
fuel_df.head()
```

```python
fuel_df.isnull().sum().head()
```

```python
fuel_df.to_csv(path+'cleaned_fuel_consumption.csv', index=False)
```
